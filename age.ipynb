```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/drive/MyDrive/age_gaps.csv")
sns.set_style("whitegrid")

# 1. Age Difference Distribution in Movies
plt.figure(figsize=(8, 6))
sns.histplot(df['age_difference'], bins=20, kde=True)
plt.title("Age Difference Distribution in Movies")
plt.xlabel("Age Difference")
plt.ylabel("Count")
plt.show()

# 2. Trend of Age Difference Over Years
plt.figure(figsize=(8, 6))
sns.lineplot(x='release_year', y='age_difference', data=df)
plt.title("Trend of Age Difference Over Years")
plt.xlabel("Release Year")
plt.ylabel("Age Difference")
plt.show()

# 3. Age Difference by Gender of Character 1
plt.figure(figsize=(8, 6))
sns.boxplot(x='character_1_gender', y='age_difference', data=df)
plt.title("Age Difference by Gender of Character 1")
plt.xlabel("Character 1 Gender")
plt.ylabel("Age Difference")
plt.show()

# 4. Actor 1 Age vs. Actor 2 Age
plt.figure(figsize=(8, 6))
sns.scatterplot(x='actor_1_age', y='actor_2_age', data=df, alpha=0.5)
plt.title("Actor 1 Age vs. Actor 2 Age")
plt.xlabel("Actor 1 Age")
plt.ylabel("Actor 2 Age")
plt.show()

# 5. Top 10 Movies with Largest Age Differences
top_movies = df.nlargest(10, 'age_difference')[['movie_name', 'age_difference']]
plt.figure(figsize=(8, 6))
sns.barplot(x=top_movies['age_difference'], y=top_movies['movie_name'])
plt.title("Top 10 Movies with Largest Age Differences")
plt.xlabel("Age Difference")
plt.ylabel("Movie Name")
plt.show()
# 6. Age of Actor 1 by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='character_1_gender', y='actor_1_age', data=df)
plt.title("Age of Actor 1 by Gender")
plt.xlabel("Character 1 Gender")
plt.ylabel("Actor 1 Age")
plt.show()

# 7. Number of Movies by Age Difference
age_gap_counts = df['age_difference'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=age_gap_counts.index, y=age_gap_counts.values)
plt.title("Number of Movies by Age Difference")
plt.xlabel("Age Difference")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.show()
```
